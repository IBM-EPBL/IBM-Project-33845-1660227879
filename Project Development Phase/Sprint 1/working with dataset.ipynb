```python
df = pd.get_dummies(df)
dummy_df = df
```

```python
dummy_df
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
dummy_df.describe(include = 'all')
```

```python
X = dummy_df.drop(['Heart Disease_Presence','Heart Disease_Absence'],axis=1)
y = dummy_df['Heart Disease_Presence']
```

```python
X.shape, y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
mms = MinMaxScaler()
X_train = mms.fit_transform(X_train)
X_test = mms.fit_transform(X_test)
```

```python
cc = pd.DataFrame(X_train)
```

```python
cc
```

```python
X_train.shape, X_test.shape
```

```python
rfm = RandomForestClassifier()
rfm.fit(X_train, y_train)
```

```python
pred = rfm.predict(X_test)
```

```python
accuracy_score(y_test, pred)
```

```python
cm = confusion_matrix(y_test,pred)
print(classification_report(y_test,pred))
```

```python
sns.heatmap(cm, annot = True, fmt = 'g', cbar = False, cmap = 'icefire', linewidths= 0.5, linecolor= 'grey')
plt.title('Confusion Matrix')
plt.ylabel('y_test')
plt.xlabel('Pred')
```
